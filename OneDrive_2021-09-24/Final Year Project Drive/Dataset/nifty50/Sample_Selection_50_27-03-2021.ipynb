```python
df=pd.read_csv('nifty50_stock_data.csv',parse_dates=['Date'],index_col='Date')
x =pd.read_csv('nifty50_stock_data.csv',parse_dates=['Date'],index_col='Date')
x = x.loc["2016-01-01" : "2020-12-31"]
#integrate the preprocessing codes. (finished)
```

```python
def number_of_years(y):        #calculates the number of years of the dataset
  p=y.index[0]                 #date of first row in the dataset (datetime format)
  q=y.index[len(y)-1]          #date of last row in the dataset  (datetime format)
  return ((q-p).days+1)/365    #the difference give the number of total days (not trading days) over the total number of years in the dataset
```

```python
trading_days=len(x)/number_of_years(x)                       #Trading days per year (automated)
x.reset_index(drop=True, inplace=True)
```

```python
returns=x.pct_change()                                       #getting the returns as percentage change
returns.fillna(0,inplace=True)   
```

```python
mean_return=returns.mean()*trading_days                      #getting the annual mean returns of all the remaining stocks in nifty50
```

```python
nifty = pd.read_csv('nifty50_index.csv',usecols=['Adj Close','Date'],parse_dates=['Date'],index_col='Date')
nifty = nifty.loc["2016-01-01" : "2020-12-31"]
nifty.reset_index(drop=True, inplace=True)                   #getting the nifty50 index close price
                                                             #adj. close price is ideal way. change it for the stocks too. (finished)
```

```python
nifty['Adj Close'].fillna(method='bfill',axis=0,inplace=True)
nifty['Adj Close'].fillna(method='ffill',axis=0,inplace=True)
```

```python
nifty_return = nifty.pct_change() 
nifty_return.fillna(0,inplace=True)
mean_nifty_return=nifty_return.mean()[0]*trading_days        #getting the annual mean returns of nifty50 index
```

```python
def rempos(x):                                               #mark positive values as 0 because semideviation only has negative returns
  for i in range(0,len(x)):
    if x['Adj Close'][i]>0:
      x['Adj Close'][i]=0
  return x
def rempo(x):
  for i in range(0,len(x)):
    if x[i]>0:
      x[i]=0
  return x
```

```python
drisk_nifty=rempos(nifty_return.copy()).std()[0]*trading_days   #downside risk i.e., semi-deviation of nifty50 index
```

```python
risk_free_rate = 0.0346                   #initializing risk free rate that will be used in calculating both the ratios (absolute value)
#referred from url: https://www.rbi.org.in/Scripts/BS_NSDPDisplay.aspx?param=4&Id=24292
#In the above url, the 364 (1 year => 01-01-2021) day treasury bill is 3.46% , when taken absolute value => 0.0346
# (improved)
```

```python
def ratio(a,b,c):                         #function to calculate ratio i.e. "(returns-(risk_free_rate))/deviation"
  return (a-c)/b                          #a => annual return, c => risk_free_rate, b => deviation (standard for sharpe, semi for sortino)
```

```python
sort_nifty=ratio(mean_nifty_return,drisk_nifty,risk_free_rate)      #getting the sortino ratio of nifty50 index
```

```python
for i in returns.columns.to_list():                                 #for every remaining stock in the data
  ret=returns[i].mean()*trading_days                                #get the annual mean return
  drisk=rempo(returns[i].copy()).std()*trading_days                 #get the semideviation
  sortino=ratio(ret,drisk,risk_free_rate)                           #get the sortiono ratio
  if sortino < sort_nifty:                                          #compare it with nifty50 index's sortino ratio
    returns.drop(labels=[i],axis=1,inplace=True)                    #if less, drop it
    x.drop(labels=[i],axis=1,inplace=True)
    df.drop(labels=[i],axis=1,inplace=True)                         #literature servey needed.
```
