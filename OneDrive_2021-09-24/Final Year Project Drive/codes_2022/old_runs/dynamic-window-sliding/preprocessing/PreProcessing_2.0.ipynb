```python
df=pd.read_csv("n50_filter1.csv",parse_dates=['Date'],index_col='Date')
d=pd.read_csv('n50_filter1.csv',parse_dates=['Date'],index_col='Date')
d=d.loc["2016-01-01":"2022-3-15"]
```

```python
def number_of_years(y):        #calculates the number of years of the dataset
  p=y.index[0]                 #date of first row in the dataset (datetime format)
  q=y.index[len(y)-1]          #date of last row in the dataset  (datetime format)
  return ((q-p).days+1)/365    #the difference give the number of total days (not trading days) over the total number of years in the dataset
```

```python
trading_days=len(d)/number_of_years(d)                       #Trading days per year (automated)
d.reset_index(drop=True, inplace=True)
```

```python
returns=d.pct_change()
returns.fillna(0,inplace=True)
#getting the returns as percentage change
```

```python
mean_return=returns.mean()*trading_days
```

```python
niftyindx = pd.read_csv('n50_1jan2016-1_mar_2022.csv',usecols=['Adj Close','Date'],parse_dates=['Date'],index_col='Date')
niftyindx = niftyindx.loc["2016-01-01" : "2021-3-1"]
niftyindx.reset_index(drop=True, inplace=True)  
```

```python
niftyindx['Adj Close'].fillna(method='bfill',axis=0,inplace=True)
niftyindx['Adj Close'].fillna(method='ffill',axis=0,inplace=True)
```

```python
niftyindx_return = niftyindx.pct_change() 
niftyindx_return.fillna(0,inplace=True)
mean_niftyindx_return=niftyindx_return.mean()[0]*trading_days        #getting the annual mean returns of nifty50 index
```

```python
def rempos(x):  #mark positive values as 0 because semideviation only has negative returns
  for i in range(0,len(x)):
    if x['Adj Close'][i]>0:
      x['Adj Close'][i]=0
  return x
def rempo(x):
  for i in range(0,len(x)):
    if x[i]>0:
      x[i]=0
  return x
```

```python
drisk_niftyindx=rempos(niftyindx_return.copy()).std()[0]*trading_days  
#downside risk i.e., semi-deviation of nifty50 index
```

```python
risk_free_rate = 0.0426                   #initializing risk free rate that will be used in calculating both the ratios (absolute value)
#referred from url: https://www.rbi.org.in/Scripts/BS_NSDPDisplay.aspx?param=4&Id=24292
#In the above url, the 364 (1 year => 01-01-2021) day treasury bill is 4.26% , when taken absolute value => 0.0426
# (improved)
```

```python
def ratio(a,b,c):#function to calculate ratio i.e. "(returns-(risk_free_rate))/deviation"
  return (a-c)/b  #a => annual return, c => risk_free_rate, b => deviation (standard for sharpe, semi for sortino)
```

```python
sortino_niftyindx=ratio(mean_niftyindx_return,drisk_niftyindx,risk_free_rate)      #getting the sortino ratio of nifty50 index
```

## Filter 2
### Nifty50 index Sortino ratio is compared with each assets sortino ratio
### if asset's srotino ratio is less than nifty index sortino ratio the asset is dropped

```python
for i in returns.columns.to_list(): #for every remaining stock in the data
  ret=returns[i].mean()*trading_days  #get the annual mean return
  drisk=rempo(returns[i].copy()).std()*trading_days #get the semideviation
  sortino=ratio(ret,drisk,risk_free_rate)   #get the sortiono ratio
  if sortino < sortino_niftyindx:  #compare it with nifty50 index's sortino ratio
    returns.drop(labels=[i],axis=1,inplace=True)    #if less, drop it
    d.drop(labels=[i],axis=1,inplace=True)
    df.drop(labels=[i],axis=1,inplace=True)       
```
